```python
data=pd.read_csv('book (2).csv',encoding='Latin-1',error_bad_lines=False)

data
```

```python
data=data.sort_values(['User.ID'])
```

```python
len(data['Book.Title'].unique())
```

```python
data_pivot=data.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
data_pivot
```

```python
data_pivot.index=data['User.ID'].unique()
data_pivot
```

```python
data_pivot.fillna(0,inplace=True)
data_pivot
```

```python
user_sim=1-pairwise_distances(data_pivot,metric='cosine')
user_sim
```

```python
similar_df=pd.DataFrame(user_sim)
similar_df
```

```python
similar_df.index=data['User.ID'].unique()
similar_df.columns=data['User.ID'].unique()
similar_df
```

```python
np.fill_diagonal(user_sim,0)
similar_df
```

```python
similar_df.idxmax(axis=1)
```
